```python
from matplotlib  import pyplot as plt

months_x = ["Jan", "Feb", "March", "Apr", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]
dev_y = [400, 800, 850, 900, 930, 940, 960, 1000, 1100, 1150, 1180, 1200]
plt.plot(months_x, dev_y, "g--", label = "in person")

t_dev_y = [800, 790, 810, 750, 700, 650, 640, 600, 550, 480, 500, 400]

plt.plot(months_x, t_dev_y, "b--", label = "by fax/email")

p_dev_y = [1000, 1020, 1050, 1030, 1200, 1250, 1300, 1320, 1280, 1350, 1400, 1420]

plt.plot(months_x, p_dev_y, "k--", label = "by telephone")

plt.title("No.of complaints to an Airline over a year")
plt.xlabel("months")
plt.ylabel("types of complaint")
plt.legend()
plt.show()
```
